```python
data[['HDL', 'LDL']] = data['Cholesterol Levels'].str.extract(r'HDL: (\d+), LDL: (\d+)').astype(int)
```

```python
cate_columns = ['Gender', 'Marital Status', 'Work Type', 'Residence Type',
                       'Alcohol Intake', 'Physical Activity', 'Family History of Stroke',
                       'Dietary Habits', 'Symptoms', 'Diagnosis']
```

```python
data = pd.get_dummies(data, columns=cate_columns, drop_first=True)
```
